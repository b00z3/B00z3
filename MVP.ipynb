```python
import ipywidgets as widgets
from IPython.display import display
```

```python
a = widgets.IntSlider(min=19,max=100,description='Age')
team = []
g = widgets.RadioButtons(
    options=['Male', 'Female', 'Other'],
#     value='pineapple',
    description='Gender:',
    disabled=False
)
m = widgets.ToggleButton(
    value=False,
    description='Married?',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Description',
    icon='check'
)
button = widgets.Button(
    description='Add member',
    disabled=False,
    button_style='info',
)
output = widgets.Output()
def on_button_clicked(b):
    with output:
        team.append(
            {
                'age':a.value,
                'gender':g.value,
                'married':m.value,
            }
        
        )
        print(avg(team))
        print("PREDICTED")
        print(knn_clf.predict([avg(team)]))
button.on_click(on_button_clicked)

display(a)
display(g)
display(m)
display(button, output)
```

```python


def ismale(male):
    if male == 'Male':
        return 1
    else:
        return 0

def avg(team):
    age=0
    gender = 0
    married = 0
    for n in team:
        age = age + n['age']
        gender = gender + ismale(n['gender'])
        married = married + n['married']
    age = age/len(team)
    gender = gender/len(team)
    married = married/len(team)
    return [age, gender, married]

avg(team)
```
