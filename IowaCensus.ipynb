```python
import pandas as pd
from sklearn import preprocessing
```

```python
df = pd.read_csv('Data/iowa_census_data.csv', encoding = "utf-16", delimiter='\t')
```

```python
df = df.drop('Year', axis=1)
```

```python
df = df.drop('Hispanic or Latino origin', axis=1)
```

```python
df['Race'].unique()
```

```python
df['Race'] = df['Race'].replace('Black and African American alone','Black')
df['Race'] = df['Race'].replace('Asian alone','Asian')
df['Race'] = df['Race'].replace('American Indian and Alaska Native alone','First Nations')
df['Race'] = df['Race'].replace('White alone', 'White')
df['Race'] = df['Race'].replace('Native Hawiian and Other Pacific Islander alone','Hawaiian or Pacific Islander')
df = df[df['Race'] != 'All races']
```

```python
df = df[df['Sex'] != 'Total population']
```

```python
df = df[df['Age group'] != 'Total population']
```

```python
df = df[df['Age group'] != '0 to 4 years of age']
df = df[df['Age group'] != '5 to 9 years of age']
df = df[df['Age group'] != '10 to 14 years of age']
df = df[df['Age group'] != '15 to 19 years of age']
```

```python
df = df.pivot_table('Data',['Age group','Sex','Race'],'County Name')
```

```python
df = df.transpose()
df.columns = [' '.join(col).strip() for col in df.columns.values]
```

```python

total = pd.Series(df.T.sum(numeric_only=True),name='Total')

```

```python
df.loc["Adair County":"Wright County",:].div(total, axis=0)
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
df = pd.DataFrame(min_max_scaler.fit_transform(df), columns=df.columns, index=df.index)
```
