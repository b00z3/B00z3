# Normalize

```python
minn = all_data['beer avg'].min()
maxx = all_data['beer avg'].max()

f = lambda x: (x - minn)/(maxx - minn)

all_data['beer avg'] = all_data['beer avg'].map(f)
```

```python
minn = all_data['wine avg'].min()
maxx = all_data['wine avg'].max()

f = lambda x: (x - minn)/(maxx - minn)

all_data['wine avg'] = all_data['wine avg'].map(f)
```

```python
minn = all_data['spirits avg'].min()
maxx = all_data['spirits avg'].max()

f = lambda x: (x - minn)/(maxx - minn)

all_data['spirits avg'] = all_data['spirits avg'].map(f)
```

```python
minn = all_data['ref avg'].min()
maxx = all_data['ref avg'].max()

f = lambda x: (x - minn)/(maxx - minn)

all_data['ref avg'] = all_data['ref avg'].map(f)
```

```python
minn = all_data['beer %'].min()
maxx = all_data['beer %'].max()

f = lambda x: (x - minn)/(maxx - minn)

all_data['beer %'] = all_data['beer %'].map(f)
```

```python
minn = all_data['wine %'].min()
maxx = all_data['wine %'].max()

f = lambda x: (x - minn)/(maxx - minn)

all_data['wine %'] = all_data['wine %'].map(f)
```

```python
minn = all_data['spirits %'].min()
maxx = all_data['spirits %'].max()

f = lambda x: (x - minn)/(maxx - minn)

all_data['spirits %'] = all_data['spirits %'].map(f)
```

```python
minn = all_data['ref %'].min()
maxx = all_data['ref %'].max()

f = lambda x: (x - minn)/(maxx - minn)

all_data['ref %'] = all_data['ref %'].map(f)
```

```python
all_data.rename(columns={ all_data.columns[0]: "male %" }, inplace = True)
```

```python
all_data.rename(columns={ all_data.columns[1]: "avg age" }, inplace = True)
```

```python
all_data.rename(columns={ all_data.columns[2]: "married %" }, inplace = True)
```

```python
minn = all_data['male %'].min()
maxx = all_data['male %'].max()

f = lambda x: (x - minn)/(maxx - minn)

all_data['male %'] = all_data['male %'].map(f)
```

```python
minn = all_data['avg age'].min()
maxx = all_data['avg age'].max()

f = lambda x: (x - minn)/(maxx - minn)

all_data['avg age'] = all_data['avg age'].map(f)
```

```python
minn = all_data['married %'].min()
maxx = all_data['married %'].max()

f = lambda x: (x - minn)/(maxx - minn)

all_data['married %'] = all_data['married %'].map(f)
```
