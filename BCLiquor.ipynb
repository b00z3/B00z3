```python
import pandas as pd
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_csv('bcliquor.csv', thousands=',')
df.head()
```

```python
df['REGIONAL DISTRICT NAME'].unique()
```

```python
df['FISCAL_YEAR_NAME'].unique()
```

```python
df['PRICE/L'] = df['RETAIL AMT ($)'].astype('float')/df['SALE VOLUME (LITRES)'].astype('float')
```

```python
df = df.pivot_table(['SALE VOLUME (LITRES)', 'RETAIL AMT ($)','PRICE/L'],['FISCAL_YEAR_NAME','REGIONAL DISTRICT NAME'],'PRODUCT CLASS NAME')

df.head()
```

```python
df = df.unstack(level=0)
```
